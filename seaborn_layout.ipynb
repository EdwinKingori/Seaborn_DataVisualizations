## Box Plot

A box plot (or box-and-whisker plot) shows the distribution of quantitative
data in a way that facilitates comparisons between variables or across
levels of a categorical variable. 
The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution,
except for points that are determined to be "outliers" using a method
that is a function of the inter-quartile range.


```python
sns.boxplot(data = tips_df, x ='day', y = 'total_bill', hue = 'sex')
plt.legend(loc = 0)
```

From the plot above: 
    The barline that goes accross he middle of the boxes is the median, and the box extends one_standard deviation(that's what the box represents) from the median.
The lines above the boxes are whiskers. And they extend to all of the other data asides from what is in the standaed deviation. The star-like shapes are considered to be outliers.

The inclusion of the hue provides us with the ability to introduce an additional category. In this specific scenario, by incorporating sex as the hue, we are able to generate comparative boxplots for each day, differentiating between males and females.

## Strip Plot

Strip plots are similar to scatter plots but have only one axis representing the data points. The data points are displayed as individual marks (such as dots or short lines) along the axis, showing their respective values. 

```python
plt.figure(figsize = (8,5))

sns.stripplot(data = tips_df, x = 'day', y = 'total_bill', jitter = True, hue = 'sex', dodge = True )
#Dodge separates the male and female data for each day. 
```

## Pair Grid

The PairGrid allows you to customize each subplot individually, making it a versatile way to visualize various types of data distributions and correlations. It is a high-level interface for creating pair plots, and it uses other Seaborn plotting functions to create the individual subplots.

The grid.map() function is used to map the type of plot you want to create on the grid. You can use any Seaborn plotting function, such as scatterplot, kdeplot, histplot, etc.

```python
iris = sns.load_dataset('iris')
iris_df = sns.PairGrid(iris, hue = 'species')
# visualizing the scatterplots diognally and in both the upper na lower sections using the plt.scatter
iris_df.map(plt.scatter)
# putting a histogram across the diognal 
iris_df.map_diag(plt.hist)
```

```python
#Placing a scatterplot in the upper grid, a kde plot in the lower grid, and a histogram in the diognal grid  
iris_df = sns.PairGrid(iris, hue = 'species')

# plotting the diognal hist
iris_df.map_diag(plt.hist)

# upper scatter and lower kde
iris_df.map_upper(plt.scatter)
iris_df.map_lower(sns.kdeplot)
```

```python
# Creating a custom PairGrid plot
iris_df = sns.PairGrid(iris, hue = 'species', 
                      x_vars = ['sepal_length', 'sepal_width'],
                      y_vars = ['petal_length', 'petal_width'])

iris_df.map(plt.scatter)
iris_df.add_legend()

```
